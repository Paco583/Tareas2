# 1.- Suma de Matices Rectangulares

```python
#Arturo Gongora
import numpy as np

M1 = np.array([[1,2,3], #Declaramos las matrices
             [4,5,6],
             [7,8,9]])

M2 = np.array([[10,11,12],
             [13,14,15],
             [16,17,18]])

def suma(M1,M2): #Definimos la función que sumará nuestras matrices.
  filas,columnas = M1.shape #Tomamos la dimensión de alguna matriz.
  MS = np.zeros((filas,columnas),dtype = int) #Creamos una nueva matriz vacía del mismo tamaño para guardar los resultados.
  for i in range(filas):  #Iteramos sobre las filas.
    for j in range(columnas):  #Iteramos sobre las columnas.
      MS[i, j] = M1[i, j] + M2[i, j] #Realizamos la suma elemento a elemento.
  return MS

MS = suma(M1,M2) #Llamamos a la función y guardamos el resultado.
print(MS)
```

# 2.- Multiplicacion entrada a entrada de matrices rectangulares

```python
#Arturo Gongora
import numpy as np

M1 = np.array([[1,2,3], #Declaramos las matrices
             [4,5,6],
             [7,8,9]])

M2 = np.array([[10,11,12],
             [13,14,15],
             [16,17,18]])

def producto(M1,M2): #Definimos la función que multiplicará nuestras matrices elemento por elemento.
  filas,columnas = M2.shape #Tomamos la dimensión de alguna matriz.
  MP = np.zeros((filas,columnas),dtype = int) #Creamos una nueva matriz vacía del mismo tamaño para guardar los resultados.
  for i in range(filas):  #Iteramos sobre las filas.
    for j in range(columnas):  #Iteramos sobre las columnas.
      MP[i, j] = M1[i, j] * M2[i, j] #Realizamos el producto elemento a elemento.
  return MP

MP = producto(M1,M2) #Llamamos a la función y guardamos el resultado.
print(MP)
```

# 3.- La traza de una matriz

```python
#Arturo Gongora
import numpy as np

M1 = np.array([[1,2,3], #Declaramos la matriz.
             [4,5,6],
             [7,8,9]])

def traza(M1): #Definimos la función que realizará la sumatoria de los elementos diagonales (necesarios para la traza).
  sumatoria = 0 #Inicializamos la sumatoria.
  for i in range(len(M1)):  #Iteramos sobre los índices diagonales.
    sumatoria = sumatoria + M1[i, i]  #Sumamos los elementos ii.
  return sumatoria

tr = traza(M1) #Llamamos a la función.
print("tr = ",tr)
```

# 4.- La transpuesta de una matriz

# 5.- Encontrar el maximo y minimo de una matriz

```python
#Francisco Padilla
import numpy as np
B = np.array([[1,2],[3,4]]) #Declaramos la matriz

a = B[0,0] #Separamos los elementos de la matriz
b = B[0,1]
c = B[1,0]
d = B[1,1]

#Buscamos el Maximo
if a>b and a>c and a>d:
  print("El maximo es",a)
elif b>a and b>c and b>d:
  print("El maximo es",b)
elif c>a and c>b and c>d:
  print("El maximo es",c)
elif d>a and d>b and d>c:
  print("El maximo es",d)

#Buscamos el minimo
if a<b and a<c and a<d:
  print("El minimo es",a)
elif b<a and b<c and b<d:
  print("El minimo es",b)
elif c<a and c<b and c<d:
  print("El minimo es",c)
elif d<a and d<b and d<c:
  print("El minimo es",d)
```

# 6.- Calcular la suma del cuadrado de los elementos de una matriz.

```python
#Arturo Gongora
import numpy as np

M1 = np.array([[1,2,3], #Declaramos la matriz.
             [4,5,6],
             [7,8,9]])

def suma_cuadrados(M1): #Definimos la función que sumará los cuadrados de cada elemento.
  filas,columnas = M1.shape
  suma_c = 0 #Inicializamos la sumatoria.
  for i in range(filas):  #Iteramos sobre las filas.
    for j in range(columnas):  #Iteramos sobre las columnas.
      suma_c = suma_c + M1[i, j]**2 #Elevamos al cuadrado los elementos y los sumamos.
  return suma_c

suma_c = suma_cuadrados(M1) #Llamamos a la función.
print(suma_c)
```

# 8.- Dadas las matrices A y B, calculara


*   2B + 3A
*   (B+2A)**2

```python
#Francisco Padilla
import numpy as np
A = np.array([[11,7,-1],[-3,5,2],[4,2,-2]]) #Declaramos la matriz A
B = np.array([[1,1,1],[10,3,-5],[-6,-10,-3]]) #Declaramos la matriz B

Z = 2*B + 3*A #Realizamos la primera operacion
X = (B+2*A)**2 #Realizamos la segunada operacion
Y = A**2 + 3*B #Realizamos la tercer operacion

#Imprimimos los resultados obtenidos
print("2B + 3A es:")
print(Z)

print("B+2A**2 es:")
print(X)

print("A**2 + 3B es:")
print(Y)
```

# 9.- Problema de clase de ecuaciones diferenciales


```python
#Arturo Gongora
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500) #Definimos un rango de valores para x y el número de puntos a graficar.
C_mayor0 = [1,2,3,4,5]  #Definimos los valores para C.
C_igual0 = [0]
C_menor0 = [-1,-2,-3,-4,-5]
plt.figure(figsize=(10, 6)) #Creamos la figura y los ejes.

for C in C_mayor0: #Graficamos para cada tipo de C utilizando for.
  plt.plot(x, (x-1) + C * np.exp(-x), color='y', linestyle='-')
for C in C_igual0:
  plt.plot(x, (x-1) + C * np.exp(-x), color='g', linestyle='dotted', linewidth=4)
for C in C_menor0:
  plt.plot(x, (x-1) + C * np.exp(-x), color='b', linestyle='-.')

plt.xlabel("Eje X") #Señalamos nuestros ejes.
plt.ylabel("Eje Y")
f1= {'family':'serif', 'color':'k', 'size':20} #Definimos f1 para caracterizar el título.
plt.title("Familia de curvas para la solución de la ecuación diferencial; [-5 </= x </= 5]", fontdict = f1)
legend_labels = ["C > 0 (Amarillo)", "C = 0 (Verde)", "C < 0 (Azul)"] #Creamos una sola leyenda que muestre los tres casos.
plt.legend(legend_labels, loc='upper right') #Mostramos la leyenda y la colocamos en la ezquina superior derecha.
plt.grid(alpha=0.5) #Controlamos la transparencia de la grilla.
plt.show() #Mostramos la gráfica.
```

#10.- Grafica de (3x+7)/(x**2-4)

```python
#Francisco Padilla
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10, 10000) #Generamos el espacio
y = (3*x+7)/(x**2 - 4) #Definimos la funcion

plt.plot(x,y, label = "y=(3x+7)/(x**2 - 4)")#"Exportamos" x, y y asignamos un identificador

plt.ylim(-10,10) #Limitamos el rango de x y y
plt.xlim(-10,10)

#Graficamos las asintotas en x=2,x=-2 y y=0
plt.axvline(x=2, color='y', linestyle='--', label='Asintota vertical x=2')
plt.axvline(x=-2, color='y', linestyle='--', label='Asintota vertical x=-2')
plt.axhline(y=0, color='g', linestyle='--', label='Asintota horizontal y=0')

plt.grid()#Habilitamos la malla o el egrid
plt.title("Grafica de y=(3x-7)/(x**2 - 4)") #Titulo
plt.xlabel("Eje x") #Leyenda en el eje x
plt.ylabel("Eje y") #Leyenda en el eje y
plt.legend(loc = "upper left", fontsize = 7) #imprimimos las leyendas y las acomodamos en la grafica

plt.show()
```
